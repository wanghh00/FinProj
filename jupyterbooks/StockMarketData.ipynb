```python
def ToMonthStr(month):
    return str(month) if month >= 10 else "0%s" % str(month)

def GenMonFromdaily(dailyDf, year, month):
    monthStr = ToMonthStr(month)
    start = '%s-%s-01' % (year, monthStr)
    end = '%s-%s-31' % (year, monthStr)
    tmp = dailyDf.loc[start:end]
    #return [tmp.iloc[0]['Open'], max(tmp.loc[:,'High']), min(tmp.loc[:,'Low']), tmp.iloc[-1]['Close'], sum(tmp.loc[:,'Volume'])]
    return [tmp.iloc[0]['Open'], tmp.iloc[0]['High'], tmp.iloc[0]['Low'], tmp.iloc[0]['Close'], tmp.iloc[0]['Volume']]
    #return [tmp.iloc[0]['Open'], tmp.iloc[0]['High'], tmp.iloc[0]['Low'], tmp.iloc[0]['Close'], 5]
```

```python
monthly_spy = pd.DataFrame(columns=['Open','High','Low','Close','Volume'])
for year in range(1993,2018):
    for month in range(1,13):
        monthly_spy.loc["%s-%s" % (year, ToMonthStr(month))] = GenMonFromdaily(raw_spy, year, month)
        
monthly_spy.loc['2018-01'] = GenMonFromdaily(raw_spy, 2018, 1)
monthly_spy.loc['2018-02'] = GenMonFromdaily(raw_spy, 2018, 2)
monthly_spy
```
